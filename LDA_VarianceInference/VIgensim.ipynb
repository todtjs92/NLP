```python
import gensim
from gensim import corpora
```

```python
new_corpus = [['1','2'],['3','4']]
```

```python
dict_ = corpora.Dictionary(new_corpus)
```

```python
dict_
```

```python

print(dict_)
```

```python
doc_term_matrix = [dict_.doc2bow(i) for i in new_corpus]
doc_term_matrix
```

```python
new_corpus
```

```python

# Creating the object for LDA model using gensim library

Lda = gensim.models.ldamodel.LdaModel
```

```python
ldamodel = Lda(doc_term_matrix, num_topics=2, id2word = dict_, alpha='auto', random_state=0, eval_every=1)
```

```python
from gensim.models import CoherenceModel
```

```python
ldamodel.get_topics()
```

```python
# Compute Coherence Score
coherence_model_lda = CoherenceModel(
    
    topics = new_corpus,
    corpus=doc_term_matrix , 
    dictionary =dict_ ,
    coherence='u_mass',
    topn=3
    )
```
